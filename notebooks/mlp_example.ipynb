```python
from tml.model.pipeline import Pipeline, ModelHandler
from tml.models.model import BinaryClassificationLightning
from tml.data.load_data import load_and_preprocess_data
from tml.model.analysis import BinaryClassificationAnalysis
```

```python
# Prepare the data
cols = range(1, 42+1)
input_dim = len(cols) - 1
input_path = '../data/test1.csv'
load_dict =load_and_preprocess_data(input_path, cols)
dataset = load_dict['test_set']
data = dataset[:, 1:]
hard_targets = dataset[:, 0]

```

```python
model_instance = BinaryClassificationLightning(input_dim=41, nb_classes=1, dropout_rate_train=0.2, dropout_rate_pred=0.7, learning_rate=0.001)

# OR

model_class = BinaryClassificationLightning
model_config = {
    "input_dim": 41,
    "nb_classes": 1,
    "dropout_rate_train": 0.5,
    "dropout_rate_pred": 0.7,
    "learning_rate": 0.001
}
```

```python
out1 = load_dict['out1']
out2_var = load_dict['out2_var']
```

```python
# Initialize the pipeline
model_handler = ModelHandler(model_instance=model_instance, model_class=model_class, model_config=model_config)
pipeline = Pipeline(model_handler, data, hard_targets, max_epochs=10, drop_iterations=100)



```

```python
# model_class(**model_config)
```

```python
pipeline.run(n_steps=25)
```

```python
pipeline.hard_targets.numpy()
```

```python
pipeline.probability_scores
```

```python
pipeline.uncertainty_scores
```

```python
# Assuming labels, probability_scores, and uncertainty_scores are available
analysis = BinaryClassificationAnalysis(pipeline.hard_targets.numpy(),
                                        pipeline.probability_scores,
                                        pipeline.uncertainty_scores)

# Calculate metrics
roc_auc = analysis.calculate_roc_auc()
brier_score = analysis.calculate_brier_score()
print(f"ROC AUC: {roc_auc}, Brier Score: {brier_score}")

# Plotting
analysis.plot_reliability_diagram()
analysis.plot_roc_curve()
analysis.plot_uncertainty_distribution()
analysis.plot_uncertainty_vs_confidence()
```
